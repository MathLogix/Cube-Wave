```python
import pygame
from pygame.locals import *
from OpenGL.GL import *
from OpenGL.GLU import *
import numpy as np
import math

w = 24
frames = 60
angle = 0
maxD = np.sqrt(200**2 + 200**2)
ma = math.atan(math.cos(math.pi / 4))

def draw_box(x, y, z, w, h):
    glPushMatrix()
    glTranslatef(x, y, z)
    glBegin(GL_QUADS)
    # Front face
    glColor3f(1, 0, 0)  # Red
    glVertex3f(-w/2, -h/2, w/2)
    glVertex3f(w/2, -h/2, w/2)
    glVertex3f(w/2, h/2, w/2)
    glVertex3f(-w/2, h/2, w/2)
    # Back face
    glColor3f(0, 1, 0)  # Green
    glVertex3f(-w/2, -h/2, -w/2)
    glVertex3f(w/2, -h/2, -w/2)
    glVertex3f(w/2, h/2, -w/2)
    glVertex3f(-w/2, h/2, -w/2)
    # Top face
    glColor3f(0, 0, 1)  # Blue
    glVertex3f(-w/2, h/2, -w/2)
    glVertex3f(w/2, h/2, -w/2)
    glVertex3f(w/2, h/2, w/2)
    glVertex3f(-w/2, h/2, w/2)
    # Bottom face
    glColor3f(1, 1, 0)  # Yellow
    glVertex3f(-w/2, -h/2, -w/2)
    glVertex3f(w/2, -h/2, -w/2)
    glVertex3f(w/2, -h/2, w/2)
    glVertex3f(-w/2, -h/2, w/2)
    # Right face
    glColor3f(1, 0, 1)  # Magenta
    glVertex3f(w/2, -h/2, -w/2)
    glVertex3f(w/2, -h/2, w/2)
    glVertex3f(w/2, h/2, w/2)
    glVertex3f(w/2, h/2, -w/2)
    # Left face
    glColor3f(0, 1, 1)  # Cyan
    glVertex3f(-w/2, -h/2, -w/2)
    glVertex3f(-w/2, -h/2, w/2)
    glVertex3f(-w/2, h/2, w/2)
    glVertex3f(-w/2, h/2, -w/2)
    glEnd()
    glPopMatrix()

def main():
    global angle
    pygame.init()
    display = (800, 800)
    pygame.display.set_mode(display, DOUBLEBUF | OPENGL)
    gluPerspective(45, (display[0] / display[1]), 0.1, 1000.0)
    glTranslatef(0.0, 0.0, -300)
    glRotatef(-90, 1, 0, 0)

    clock = pygame.time.Clock()

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()

        glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
        glPushMatrix()
        glRotatef(-math.degrees(ma), 1, 0, 0)
        glRotatef(-45, 0, 1, 0)

        for z in range(0, 400, w):
            for x in range(0, 400, w):
                d = np.sqrt((x - 200)**2 + (z - 200)**2)
                offset = np.interp(d, [0, maxD], [-np.pi, np.pi])
                a = angle + offset
                h = int(np.interp(np.sin(a), [-1, 1], [50, 150]))
                draw_box(x - 200, 0, z - 200, w, h)

        angle -= 2 * np.pi / frames
        glPopMatrix()
        pygame.display.flip()
        clock.tick(30)

if __name__ == "__main__":
    main()
```
